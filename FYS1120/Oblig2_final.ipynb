# Oblig 1

Vi ønsker å implementere kabel likningen numerisk. Vi skal starte først med en enkel modell for en enhet før vi gjør ting mer komplisert med en lang rekke av disse enhetene. For å få en bedre forståelse for hvordan en enhet oppfører seg skal vi se på ulike mulige oppbygninger av enheten. Hver enhet består av en krets med en kapasitanse parallellkoblet med en komponent $X$, som vi vil endre på. Vi vil introdusere et pulssignal for å se på hva vi får ut på andre enden. Vi vil så modifisere modellen, altså endre på komponent $X$, for en bedre tilnærming til det systemet vi studerer. Systemet vi ønsker å studere er flyten av ioner igjennom en cellemembran.
Her ser du en systemet vi ønsker å modulere, hvor hver boks er en enhet.

## Likninger
For å beregne hvordan pulsene beveger seg igjennom systemet trenger vi likninger for hvordan hver modul oppfører seg. <br>

Vi ser at strømmen som går inn i hver model kan skrives som, ved hjelp av Kirchoffs strøm lov i likning under. <br>
\begin{equation}
    i_{j-1}=i_j+I_{C, j}+I_{X, j}
\end{equation}

Mellom hver modul har vi har vi en resistans $r$ på insiden av cellen. Vi har at vi kan da skrive strømmen inn i modulen som potensial forskjellen mellom de to modulen delt på denne resistansen $r$, altså:
\begin{equation}
    i_{j-1}=\frac{V_{j-1}-V_j}{r}
\end{equation}
Vi kan gjøre det samme for $i_j$.<br>

Vi kan skrive strømmen igjennom en kondensator ved $I=\frac{dQ}{dt}$. Setter vi da inn at $Q=CV$, hvor $C$ er den konstante kapasitansen til kondensatoren, og $V$ er potensial forskjellen. Ettersom potensialet på utsiden alltid er null står vi igjen med utrykket under for strømmen $I_{C,j}$.
\begin{equation}
    I_{C, j}=C \frac{dV_j}{dt}
\end{equation}
Ettersom vi ikke enda har bestemt hva X er, får vi likningen under, etter litt triksin, til å beskrive endringen i potensialet.
\begin{equation}
    C\frac{dV_j}{dt}=\frac{V_{j+1}+2V_j+V_{j-1}}{r}-I_{X,j}
\end{equation}
For den første modulen, altså j$=1$, erstatter vi $V_0$ med pulsen vår $V(t)$. <br>
Det andre spesial tilfellet blir den siste modulen, altså $j=N$, hvor Kirchoffs strøm lov gir oss heller:
\begin{equation}
    i_{N-1}=I_{C, N}+I_{X, N}
\end{equation}
Som gir oss likningen for endring i potensial:
\begin{equation}
    C\frac{dV_N}{dt}=\frac{-V_N+V_{N-1}}{r}-I_{X, N}
\end{equation}
Med alle likningene våres klare, kan vi gå videre til Task 1.

## Parametere
Realistiske parametere for celle-systemet vi ønsker å studere er $C=10^{-10}F$, $R=10^{11}\Omega$, $r=10^6\Omega$ om $V_0=100mV$. C, R og r er konstanter, men $V_0$ er kun den maksimale volten vi får ut av $V(t)$. Avstanden mellom hver modul er $1\mu m$. <br>
For at vi skal kun simulere systemet i en liten periode $T=0.1s$.

## Kode
Her er koden vi bruker for å simulere systemet. Her er $L=N$, som vi kan sette selv, og $V_s$ er funksjonen for pulsen vår.

```python
import numpy as np
import matplotlib.pyplot as plt
# Setting realistic paramters
C = 1e-10 #Capacitance
R = 1e11
r = 1e6
V0 = 100e-3
T = 0.1 # Simulation in seconds
t0 = T*0.001 # Pulse length
dt = 0.1*C*r #time step
nsteps = int(T/dt) # Number of time steps

def simulate(Vs, L):
    V = np.zeros((nsteps,L),float)
    V[0,0] = V0 # Initial condition, t = 0
    t = np.zeros((nsteps,1),float)
    t_end = 0 #antall tidssteg teller
    for j in range(0,nsteps-1): #j is the timestep nr.
        if V[j, -1] == 0:
            t_end += 1 #plusser på et tidssteg
        t[j+1] = t[j] + dt
        V[j+1,0] = Vs(t[j+1],V0,t0) # Left boundary condition (i=0)
        for i in range(1,L-1): # Internal elements and i is the j-th element in the sequence
            V[j+1,i]=V[j,i]+(dt/C)*((V[j,i+1]-2*V[j,i]+V[j,i-1])/r-V[j,i]/R)
        # Right boundary condition (i=L-1)
        V[j+1,L-1] = V[j,L-1] + (dt/C)*((-V[j,L-1]+V[j,L-2])/r - V[j,L-1]/R)
        
    return V, t, t_end
```

## Task 1
Om nevnt, vil vi starte enkelt før vi utvider modellen vår. Derfor velger vi $N=1$. Samtidig starter vi med å se på komponent $X$ som en resistans $R$. Vi starter systemet uten noe volt kilde og ved tid $t=0$ skrur vi en konstant strøm $V_s$ på.<br>
Vi kan da skrive uttrykket vårt for endring i potensialet, ved hjelp av likningen vi hadde for $j=N$ og sette $V_0=V_s$.
\begin{equation}
    C\frac{dV_1}{dt}=\frac{-V_1+V_s}{r}-\frac{V_1}{R}
\end{equation}
Først ser vi på rett etter pulsen er skrudd på. Siden $V_1=0$ først vil den vokse og vokse. Siden hvor mye den vokser er avhengig av potensialet til $V_1$ et tidssteg tidligere vokser den ganske raskt først. <br>
Deretter ser vi på en stund senere. $V_1$ begynner å bli større og den vokser derfor ikke like raskt, den nærmer seg et likevekts tilstand.
Til slutt  har $V_1$ blitt ganske stor og er omtrent like stor som $V_0$. På grunn $I_{R, 1}$ leddet vil den ikke stagnisere seg ved $V_1=V_0=V_s$ I stedenfor vil den heller stagnisere seg ved:
\begin{equation}
    \frac{V_1}{R}=\frac{-V_1+V_s}{r}
\end{equation}
Fra når dette skjer, så har vil det ikke være noe endring før vi endrer på $V_s$.

## Task 2

Vi gjør modellen litt mer virkelighetsnært, og dermed litt mer komplisert, ved å øke N til 100 og sette $V_s$ til å være en puls istedenfor konstant. Jeg har valgt at pulsen skal ha formen til en sagtann for å simulere en gradvis økning av smertesignal hvor den plutselig stopper når personen fjerner f.eks. fingeren fra det som er blir varmt. Vi kan skrive denne pulsen som:
\begin{equation}
    V_s(t)=\left\{
    \begin{array}{ll}
        0 & t<0\\
        V_0\frac{t}{t_0} & 0\leq t\leq t_0\\
        0 & t>t_0
    \end{array}
    \right.
\end{equation}
Vi kan skrive den som:

```python
#Sawtooth pulse
def Vs(t, V0, t0):
    if 0<=t and t<=t0:
        return V0*t/t0
    else:
        return 0
```

Vi ønsker nå å endre på $X$ komponenten i modulene. Når vi gjør dette må vi endre på funksjonen vår for å simulere systemet vårt.<br>

Istedenfor kun en resistor $R$, introduserer vi et batteri $E$ i serie kobling med resistoren $R$. Vi setter $E=20mV$. Dette betyr at vi får et nytt uttrykk for $I_{X, j}$:
\begin{equation}
    I_{X, j} = \frac{E}{R}
\end{equation}
Dette er fordi den ene enden av batteriet alltid har det samme potensialet. <br>
En slik komponent X for cellen kan representere et konstant utslipp av ioner, uavhengig av forhold.

## Task 5
I vårt forrige system hadde vi at  potensialet ville øke raskt i starten, så avta etterhvert, for så å stabilisere seg. Dette er ikke lengre tilfellet. Vi starter med å sette potensialet på insiden $V_1=0$, før vi slipper ut den samme pulsen konstante $V_s$ <br>
Dette gir oss likningen:
\begin{equation}
    C\frac{dV_1}{dt}=\frac{-V_1+V_s}{r}-\frac{E}{R}
\end{equation}
Ved starten, altså $dV_1=0$, har vi likningen:
\begin{equation}
    C\frac{dV_1}{dt}=\frac{V_s}{r}-\frac{E}{R}
\end{equation}
Dersom $V_s/r$ er mindre enn $E/R$ vil potensialet $V_1$ først falle. Dersom det er større, vil potensialet øke.<br>
Først ser vi på hva som skjer dersom $V_s/r<E/R$. Siden potensialet $V_1$ er nå negativ vil det første leddet vokse, fordi vi har $-V_1$ som blir et positivt tall, mens det andre leddet forblir konstant. I likhet med vårt første system altså, vil endringen i starten være større enn den senere. <br>
Men hva om $V_s/r>E/R$. Da vil vi ha noe av det samme som vi hadde i vårt første system. Forskjellen her blir da at det andre leddet i likningen er konstant, og derfor mye større fra starten av, som vil bety en mer gradvis økning av potensial. <br>
Etter en god stund vil vi igjen nærme oss en likevekt:
\begin{equation}
    \frac{E}{R}=\frac{-V_1+V_s}{r}
\end{equation}
Avhengig av om $V_s/r$ er større eller mindre enn $E/R$, vil vi kunne få at potensialet stabiliserer seg ved en negativ eller positiv verdi. Dersom de er like, vil det bety at $V_1=0$ er likevekten. Det vil kunne ta lengre tid for at systemet skal stabilisere seg, avhengig av forholdene mellom de to størrelsene.
