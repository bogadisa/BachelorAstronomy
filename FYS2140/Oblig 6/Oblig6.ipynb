#### a)
Vi har at
\begin{equation}
    \psi_2=\frac{1}{\sqrt{2}}\hat{a}_+\hat{a}_+\psi_0
\end{equation}
Hvor $\hat{a}_+=\frac{1}{\sqrt{2\hbar m\omega}}(-i\hat{p}+m\omega x)$.

\begin{equation}
    \begin{split}
        \psi_1&=\frac{1}{\sqrt{2\hbar m\omega}}\left(-\hbar \frac{d}{dx}+m\omega x\right)\left(\frac{m\omega}{\pi\hbar}\right)^{\frac{1}{4}}e^{-\frac{m\omega}{2\hbar}x^2}\\
        \psi_1&=\frac{1}{\sqrt{2\hbar m\omega}}\left(\hbar\frac{m\omega}{2\hbar}2x+m\omega x\right)\left(\frac{m\omega}{\pi\hbar}\right)^{\frac{1}{4}}e^{-\frac{m\omega}{2\hbar}x^2}\\
        \psi_1&=\frac{1}{\sqrt{2\hbar m\omega}}\left(2m\omega x\right)\left(\frac{m\omega}{\pi\hbar}\right)^{\frac{1}{4}}e^{-\frac{m\omega}{2\hbar}x^2}
    \end{split}
\end{equation}

Vi gjentar prossesen for $\psi_2$, bare her må vi passe på den nye x'en.

\begin{equation}
    \begin{split}
        \psi_2&=\frac{1}{2\hbar m\omega}\left(-\hbar \frac{d}{dx}+m\omega x\right)
\left(2m\omega x\right)\left(\frac{m\omega}{\pi\hbar}\right)^{\frac{1}{4}}e^{-\frac{m\omega}{2\hbar}x^2}\\
        \psi_2&=\frac{1}{2\hbar m\omega}\left(-2m\omega\hbar+m^2\omega^2 x^2+m\omega x\right)\left(\frac{m\omega}{\pi\hbar}\right)^{\frac{1}{4}}e^{-\frac{m\omega}{2\hbar}x^2}
    \end{split}
\end{equation}

#### b)
Bruker python til å tegne:

```python
import numpy as np
from scipy.constants import hbar
import matplotlib.pyplot as plt

m = 1

omega = hbar * m * 1e-9

hbaromegam = hbar*omega*m

def psi(x, n=0):
    if n == 0:
        return (m*omega/(np.pi*hbar))**0.25*np.exp(-m*omega/(2*hbar)*x**2)
    elif n == 1:
        return 1/np.sqrt(2*hbaromegam)*(2*hbaromegam/hbar*x)*psi(x)
    elif n == 2:
        return 1/(2*hbaromegam)*(-hbaromegam + (m*omega*x)**2 + m*omega*x)*psi(x)
    else:
        print("Not supported")

N = 1000
x = np.linspace(-1, 1, N)
```

```python
fig, ax = plt.subplots(1, 3, figsize=(13,6))
for i in range(3):
    ax[i].plot(x, psi(x, i))
```
