## Oppgave 1: Aliasing

Vi har funksjonen:
\begin{equation}
    g(t)=A\sin{(2\pi ft)}
\end{equation}
med $A=1.0m$

### 1a)
Vi bruker $f=200Hz$ med sampling frekvens $f_s=1kHz$ og sampling tid $T=1s$

```python
import numpy as np
import matplotlib.pyplot as plt

A = 1 #m
f = 200
fs = 1000 #Hz
T = 1 #s

def x_n(t, f):
    return A*np.sin(2*np.pi*f*t)

N = T*fs
t = np.linspace(0, T, N)

X_k = np.fft.fft(x_n(t, f))/N

freq = np.fft.fftfreq(N, 1/fs)

plt.subplot(1,2, 1)
plt.plot(t, X_k)
plt.xlabel("t[s]")
plt.ylabel("$X_k$")
plt.subplot(1,2, 2)
plt.bar(freq, abs(X_k))
plt.xlabel("Hz")
plt.ylabel("$|X_k|$")
plt.ylim(0, 0.08)
```

### 1b)

```python
f = np.array([800, 1400, 1800])

fig, ax = plt.subplots(2, len(f), figsize=(13,6))

for i in range(len(f)):
    
    X_k = np.fft.fft(x_n(t, f[i]))/N
    freq = np.fft.fftfreq(N, 1/fs)
    
    ax[0, i].plot(t, X_k)
    ax[1, i].bar(freq, abs(X_k))
    ax[0, i].set_xlabel("t[s]")
    ax[1, i].set_xlabel("Hz")
    ax[1, i].set_ylim(0, 0.12)
```

I plottet til venstre observerer vi frekvenser rundt 200Hz. Det betyr at vi kan ungå aliasing for en samplingsfrekvens større enn ca 400Hz. Dette har med Nyquist-frekvensen å gjøre, som forteller oss at vi kan finne all informasjonen vi trenger innen for halvparten av samplingsfrekvensen. Vi finner de negative frekvensene ved hjelp av folding.
Det midterste plottet viser frekvenser rundt 400Hz, som gir minimum samplingsfrekvens 800Hz.
TIl høyre ser vi et identisk plott som det til venstre. Dette er på grunn av aliasing.

## Oppgave 2: Waveletanalyse

### 2a)

```python
A1 = 1.0 #m
A2 = 1.7 #m
f1 = 100 #Hz
f2 = 160 #Hz
t1 = 0.2 #s
t2 = 0.6 #s
sigma1 = 0.05 #s
sigma2 = 0.10 #s

def x_n(t):
    x_n1 = A1*np.sin(2*np.pi*f1*t)*np.exp(-((t-t1)/sigma1)**2)
    x_n2 = A2*np.sin(2*np.pi*f2*t)*np.exp(-((t-t2)/sigma2)**2)
    return x_n1 + x_n2

X_k = np.fft.fft(x_n(t))/N
freq = np.fft.fftfreq(N, 1/fs)

fig, ax = plt.subplots(2, 1, figsize=(13,6))
fig.tight_layout(pad=3)
ax[0].plot(t, x_n(t), color='blue', linestyle='solid', marker='o')
ax[0].set_title("$X_n$")
ax[0].set_xlabel('t [s]')
ax[0].set_ylabel('$x_n$')

ax[1].bar(freq, abs(X_k))
ax[1].set_title("FFT")
ax[1].set_xlabel('Hz')
ax[1].set_ylabel('$|x_n|$')
```

### 2b)

```python
T = 1 #s

N = int(T*fs)
t = np.linspace(0, T, N)

C = 1 #konstant

def wavelet(tn, omega_a, tk, K):
    return C*(np.exp(+1j*omega_a*(tn - tk)) - np.exp(-K**2)) * np.exp(-omega_a**2*(tn - tk)**2/(2*K)**2)

def wavelet_transform(t, signal, omega_a, K):
    gamma = np.zeros(len(t), dtype=np.complex_)
    
    s = 0
    for i, tk in enumerate(t):
        s = wavelet(t, omega_a, tk, K)
        gamma[i] = np.sum(signal*s)
        
    return gamma

omega_a1 = 80 #Hz
omega_a2 = 200 #Hz

omega_range = np.logspace(np.log10(omega_a1), np.log10(omega_a2), 100) * 2*np.pi

K1 = 6
K2 = 60
```

```python
def wavelet_diagram(t, signal, omega_range, K):
    transformed_signals = [] 
    for omega in omega_range:
        transformed_signals.append(wavelet_transform(t, signal, omega, K))
    return np.asarray(transformed_signals) # returnerer et 2D diagram for en gitt K

waveletDiagramK = []
waveletDiagramK.append(wavelet_diagram(t, x_n(t), omega_range, K1))
waveletDiagramK.append(wavelet_diagram(t, x_n(t), omega_range, K2))
```

```python
fig, ax = plt.subplots(1, 2, figsize=(13,6))
for i in range(len(ax)):
    ax[i].pcolormesh(t, omega_range / 2.0 / np.pi, np.absolute(waveletDiagramK[i]), shading='auto') 
plt.show()
```

Fra FFT plottet over i 3a) ser vi to topper, en på ca. 16oHz og en annen på 100Hz. I plottet over til høyre ser vi dette representert, bare nå over tid. De to plottene over viser sannsynligheten for hvordan bølgene oppfører seg. Til venstre har vi lav $K=6$ som gir oss mer nøyaktighet i tid, mens til høyre har vi høy $K=60$ som gir oss nøyaktighet i frekvens men usikkerhet over tid. 

## Oppgave 3: Waveletanalyse av duetrostsang

### 3a)

```python
from scipy.io import wavfile
import numpy as np
import matplotlib.pyplot as plt

fs, data = wavfile.read("mistle_thrush.wav")

x_n = data[:, 0]

N = len(x_n)
T = N/fs
t = np.linspace(0, T, N)

X_k = np.fft.fft(x_n/N)
freq = np.fft.fftfreq(N, 1/fs)
```

```python
fig, ax = plt.subplots(2, 1, figsize=(13,6))
ax[0].plot(t, x_n, color='blue')
ax[0].set_xlabel('t [s]')
ax[0].set_ylabel('x_n')
ax[1].bar(freq, abs(X_k))
ax[1].set_ylim(0, 0.0075)
```

### 3b)

```python
print(f"Samplingraten til duetrostsangen er {fs}Hz")
```

Denne er valgt fordi den gir en Nyquist-frekvens på 24kHz som betyr at den kan oppfatte de høyeste tonene mennesket kan oppfatte og dermed fremstiller den lyden mer realistisk.
