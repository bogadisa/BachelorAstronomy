```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.fft import fft, ifft
import scipy.signal
from scipy.io import wavfile
```

```python
fs, data = wavfile.read("radio-chatter_F_minor.wav")

x_n = data

N = len(x_n)
T = N/fs
t = np.linspace(0, T, N)

X_k = np.fft.fft(x_n/N)
freq = np.fft.fftfreq(N, 1/fs)
```

```python
start = int(6.9*fs)
end = int(11*fs)
x_n = data[start:end]

N = len(x_n)
T = N/fs
t = np.linspace(start/fs, end/fs, N)

X_k = np.fft.fft(x_n/N)
freq = np.fft.fftfreq(N, 1/fs)
```

```python
plt.plot(t, x_n)
plt.xlabel('t [s]')
plt.ylabel('x_n')
```

Enkelt runnning average filter

```python
def running_mean(x, windowSize):
    cumsum = np.cumsum(np.insert(x, 0, 0)) 
    return (cumsum[windowSize:] - cumsum[:-windowSize]) / windowSize 
```

Vi nsker  fjerne den konstante bakgrunnstyen fra lydklippet over. For  gjre dette vil jeg innfre et low-pass filter, noe `running_mean` er. Funksjonen er avhengig av `windowSize`, s vi m finne ut hva det betyr frst. `windowSize`er gitt ved $N=\frac{\sqrt{0.196201 + f_r^2}}{f_r}$, hvor $f_r=\frac{f_c}{f_s}$. Samplerate $f_s$ er gitt med lydfilen, men $f_c$ er noe vi m prve oss frem med. Ideen bak $f_c$ er vist under:


S vi trenger alts  beregne forholdet mellom en utgangspunktfrekvens og cutoff frekvensen som gir oss $-3dB$. S hva er denne utgangspunktfrekvensen?

Finner gjennomsnittvolum $V$ i lydfilen ved hjelp av root mean square:
\begin{equation}
    V = \sqrt{\frac{\sum_{i=0}^N x_i^2}{N}}
\end{equation}
Hvor $N$ er antall elementer i arrayen $x$.

```python
plt.figure(figsize=(12,12))
plt.subplot(2, 1, 1)
plt.plot(freq, abs(X_k))
plt.subplot(2, 1, 2)
plt.plot(freq[(0 < freq) & (freq < 4800)], abs(X_k[(0 <freq) & (freq < 4800)]))
```

```python
def V(x, freq, cutoff):
    return np.sqrt(np.sum(x[(-cutoff<=freq) & (freq<=cutoff)]**2)/len(x[(-cutoff<=freq) & (freq<=cutoff)]))
```

```python
V0 = V(X_k, freq, max(abs(freq)))
```

Vi er ute etter $-3dB$ gitt ved $-3dB=20dB\cdot \log{\left(\frac{V_0}{V_i}\right)}$, som kan skrives om til:
\begin{equation}
    V_i = 10^{\frac{3dB}{20dB}}\cdot V_0
\end{equation}
Nr vi har en verdi $V_i$ kan vi beregne $V$ med $x$ innenfor en mindre og mindre frekvens helt til vi er nrme nok. Nr vi har funnet en frekvens vi er fornyd med har vi funnet vr cut-off frekvens.

```python
Vi = 10**(3/20)*V0
f = max(freq)
tol = 1e-2
while (abs(Vi - V(X_k, freq, f)) > tol) and f > 0:
    f -= 1

cutOffFrequency = f
freqRatio = cutOffFrequency / fs

N = int(np.sqrt(0.196201 + freqRatio**2) / freqRatio)

filtered = running_mean(x_n, N)
t = np.linspace(start/fs, end/fs, len(filtered))
plt.plot(t, filtered)
plt.xlabel('t [s]')
plt.ylabel('x_n')
wavfile.write(f"Roger_lowpass{cutOffFrequency}.wav", fs, filtered.astype("float32"))
```
