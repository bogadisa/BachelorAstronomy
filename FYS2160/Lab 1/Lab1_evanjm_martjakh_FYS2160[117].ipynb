## Abstract/Introduction 

In this report we will record data from four different experiments and compare our experimental results with the theoretical results. Our goal is to find the speed of sound experimentally by measuring frequencies sent through tubes filled with different gases, and at the same time, calculate the relative error for our experiments. ## Method

   We are using standing waves to calculate the speed of sound. Three of four of the tubes are filled with air, and one filled with Argon. There is plugs at the end all the tubes and at one end ther is a small hole where a speaker is placed which sends out sound. The speaker is driven by an alternating harmonic current deliverd by a signal generator which we can change the frequency on. The plug on the other end is fitted with a microphone which is connected to a amplifier. 

   The Argon filled tube is at 20C, and the air filled tubes K2, K3, and K4 are all at different tempuratures, 20C, 70C, and 50C respectively. We measured the frequency at a range between 100-2000 Hz

In order to find the relative uncertainty, we use pythagorean error propogation:

$$\frac{df}{f} = \sqrt{{\frac{dx}{x}}^2 + {\frac{dy}{y}}^2}$$

Which in our case, can be reformulated tom 

$$ c_{id}(T) = \sqrt{\frac{(f+2)RT}{fM_{mol}}}$$

Expression for $C_v$

$$ C_{v} = \frac{dU}{dT} = \frac{d}{dT}(\frac{NfkT}{2}) = \frac{Nfk}{2} = \frac{3}{2}nR$$

since we are working with ideal gass we only have 3 degrees of freedom

Expression for $C_p$

$$ C_{p} = C_v + Nk = C_v + nR$$

The expression for $C_p - C_v$

$$ C_p - C_v = \frac{3}{2} nR + nR -  \frac{3}{2}nR = nR$$


When it comes to the MD part of our experiment, for our first simulation, we just ran the given LAMMPS file as it was given. On our second simulation, we changed the file so that the temperature starts at 0.69, which is the triple point temperature, and made it so that it increases over time. In our third simulation we changed the temperature to 1.4, its critical point, and changed the pressure to several different values.


## Results

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

```python
# temp = 20c
# K2

ResPoints = [138, 285, 417, 561, 699, 837, 976, 1115, 1254, 1394, 1532, 1673, 1811, 1950]
x = np.arange(len(ResPoints))
y = stats.linregress(x, ResPoints)


plt.plot(x, ResPoints, 'o')
plt.plot(x, y.intercept + y.slope*x)
plt.title('K2: Freq of Resonance Points of Air at Room Temp')
plt.xlabel('Resonance Points')
plt.ylabel('Frequency [Hz]')
plt.show


rel_error = np.sqrt((y.stderr/y.slope)**2 + (1.5/1243)**2)
print(f'This is relative uncertainty : {rel_error}')

exp_c = 2*1.243*y.slope
print(f'This is experimetal speed of sound : {exp_c}')

R = 8.31445
T = 293.15
M_mol = 0.0289647
f = 5

theo_c = np.sqrt(((f + 2)*R*T)/(f * M_mol)) 
print(f'This is theoretical speed of sound : {theo_c}')
```

## MD Results


### Task 2

```python

import lammps_logfile 
log = lammps_logfile.File("./log.lammps")

T = log.get("Temp")
E = log.get("TotEng")
P = log.get("Press")

# finding C_v
plt.plot(T, E)
plt.title('Energy vs. Temperature')
plt.xlabel('Temperature')
plt.ylabel('Total Energy')
plt.show()
lingress_Cv = stats.linregress(T, E)
C_v = lingress_Cv.slope
print(f'Heat Capacity [C_v] = {C_v}')

# finding Z
plt.plot(T, P)
plt.title('Pressure vs. Temperature')
plt.xlabel('Temperature')
plt.ylabel('Pressure')
plt.show()

rho = 0.001
lingress_Zrho = stats.linregress(T, P)
Z = lingress_Zrho.slope / rho # compressablity
print(f'Compressability [Z] = {Z}')

d_o_freedom = 2*C_v
print(f'nr of degrees of freedom :{d_o_freedom}')
```

The compressability is, as expected, near 1, because it behaves an ideal gas and at the same time, no simulation is perfect. Again, as expected, we get 3 degrees of freedom which matches the behavior of an ideel gass

### Task 5

```python
N = 9
C_v_array = np.zeros(N)
Z_array = np.zeros(N)
rho_list = [0.001, 0.003, 0.008, 0.013, 0.021, 0.051, 0.12, 0.3, 0.86]
for i in range(N):
    log = lammps_logfile.File(f"./log_task5_{i+1}.lammps")

    T = log.get("Temp")
    E = log.get("TotEng")
    P = log.get("Press")

    # finding C_v
    lingress_Cv = stats.linregress(T, E)
    C_v = lingress_Cv.slope
    C_v_array[i] = C_v
    print(C_v)
    

    # finding Z

    rho = rho_list[i]
    lingress_Zrho = stats.linregress(T, P)
    Z = lingress_Zrho.slope / rho # compressablity
    Z_array[i] = Z
    print(Z)

    
plt.plot(rho_list, C_v_array, 'o')
plt.title('Rho vs. C_v')
plt.xlabel('Rho')
plt.ylabel('C_v')
plt.show()

plt.plot(rho_list, Z_array, 'o')
plt.title('Rho vs. Z')
plt.xlabel('Rho')
plt.ylabel('Z')
plt.show()

```
